## delete columns that are not needed

```python
df3 = df2.drop(columns=['customer_name', 'country/region', 'city', 'state', 'postal_code', 'region', 'category', 'sub-category', 'product_id', 'sales', 'quantity', 'discount', 'profit'])
display(df3.head())
display(df3.info())
```

## Replace Whitespace with _ and upper case letters with lower case letters in rows

```python
df3['ship_mode'] = df3['ship_mode'].str.replace(' ', '_')
df3['ship_mode'] = df3['ship_mode'].str.lower()

df3['order_process_data_ship_mode'] = df3['order_process_data_ship_mode'].str.replace(' ', '_')
df3['order_process_data_ship_mode'] = df3['order_process_data_ship_mode'].str.lower()
df3.head()
```

## change date from object to datetime-format

```python
df3['order_date'] = pd.to_datetime(df3['order_date'], format="%d/%m/%Y")
df3['campaign_data_arrival_scan_date'] = pd.to_datetime(df3['campaign_data_arrival_scan_date'], format="%d/%m/%Y")
df3['order_process_data_order_date'] = pd.to_datetime(df3['order_process_data_order_date'], format="%d/%m/%Y")
df3['order_process_data_on_truck_scan_date'] = pd.to_datetime(df3['order_process_data_on_truck_scan_date'], format="%d/%m/%Y")
df3['interndata_study_ready_to_ship_date'] = pd.to_datetime(df3['interndata_study_ready_to_ship_date'], format="%d/%m/%Y")
df3['interndata_study_pickup_date'] = pd.to_datetime(df3['interndata_study_pickup_date'], format="%d/%m/%Y")
display(df3.info())
display(df3.head())
```

## change first_class to express and standard_class + second_class to standard_processing

```python
df3['order_process_data_ship_mode'] = np.where(df3['order_process_data_ship_mode'] == 'first_class', 'express', df3['order_process_data_ship_mode'])
df3['order_process_data_ship_mode'] = np.where(df3['order_process_data_ship_mode'] == 'standard_class', 'standard_processing', df3['order_process_data_ship_mode'])
df3['order_process_data_ship_mode'] = np.where(df3['order_process_data_ship_mode'] == 'second_class', 'standard_processing', df3['order_process_data_ship_mode'])
display(df3['order_process_data_ship_mode'].value_counts())
```

## Descriptive statistics

```python
df3_oc = pd.pivot_table(df3 ,index=["origin_channel"],values=["warehouse_days"],aggfunc=[np.mean, np.median, np.max])
round(df3_oc,2)
```

```python
df3_oc.describe()
```

```python
mean_values = [df3['warehouse_days'].mean(), df3['warehouse_days'].median(), df3['warehouse_days'].max()]
variance = [df3['warehouse_days'].mean() * 0.25, df3['warehouse_days'].mean() * 0.25, df3['warehouse_days'].mean() * 0.25]

bar_labels = ['Mean Score', 'Median Score', 'Max Score']
x_pos = list(range(len(bar_labels)))

plt.bar(x_pos,
        mean_values, 
        yerr=variance, 
        align='center',
        color='#FFC222',
        alpha=0.5)

plt.grid()

max_y = max(zip(mean_values, variance))
plt.ylim([0, (max_y[0] + max_y[1]) * 1.1])

plt.ylabel('Days')
plt.xticks(x_pos, bar_labels)
plt.title('Day Scores For Each Test')

plt.show()
```
